```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
```

```python
def generate_input(name, gray=False):
    image = Image.open('images/' + name + '.jpg')
    if gray:
        image = ImageOps.grayscale(image)
    image_pix = np.asarray(image)
    return image_pix, image_pix.shape
```

```python
def mul_plots(image_arrs, titles):
    fig, axs = plt.subplots(1,3, figsize=(10,7))
    for i in range(len(axs)):
        axs[i].imshow(image_arrs[i], cmap=plt.cm.binary)
        axs[i].axis('off')
        axs[i].set(title=titles[i])
        
    fig, axs = plt.subplots(1,3, figsize=(10,7))
    for i in range(len(axs)):
        axs[i].imshow(image_arrs[i+3], cmap=plt.cm.gray)
        axs[i].axis('off')
        axs[i].set(title=titles[i+3])
```

```python
def plot_boundary(name):
    # Getting the images for local storage and storing them in numpy arrays.
    input_arr, input_shape = generate_input(name)
    gray_input_arr, gray_input_shape = generate_input(name, gray=True)
    
    # Converting the numpy arrays into tensors for better computation
    input_tensor = tf.constant(input_arr, dtype='float32')
    gray_input_tensor = tf.constant(gray_input_arr, dtype='float32')
    
    # Defining the two kernels
    vertical_kernel_2d = np.array([[1,0,-1],
                                   [1,0,-1],
                                   [1,0,-1]], dtype='float32')
    
    horizontal_kernel_2d = np.array([[1,1,1],
                                     [0,0,0],
                                     [-1,-1,-1]], dtype='float32')
    
    # Expanding the kernels into 3d
    vertical_kernel_3d = np.repeat(vertical_kernel_2d[:,:,np.newaxis], 3, axis=2)
    horizontal_kernel_3d = np.repeat(horizontal_kernel_2d[:,:,np.newaxis], 3, axis=2)
    
    # Converting the kernels into tensorflow tensors
    vertical_kernel_3d = tf.constant(vertical_kernel_3d, dtype='float32')
    horizontal_kernel_3d = tf.constant(horizontal_kernel_3d, dtype='float32')
    
    # Applying the convolutions operation onto the input arrays
    clr_vert_output = tf.nn.conv3d(
        input=tf.reshape(input_tensor, [1,input_shape[2],input_shape[0],input_shape[1],1]),
        filters=tf.reshape(vertical_kernel_3d, [3,3,3,1,1]),
        strides=[1,1,1,1,1],
        padding='SAME'
    )
    clr_hori_output = tf.nn.conv3d(
        input=tf.reshape(input_tensor,  [1,input_shape[2],input_shape[0],input_shape[1],1]),
        filters=tf.reshape(horizontal_kernel_3d, [3,3,3,1,1]),
        strides=[1,1,1,1,1],
        padding='SAME'
    )
    gray_vert_output = tf.nn.conv2d(
        input=tf.reshape(gray_input_tensor, [1,gray_input_shape[0],gray_input_shape[1],1]),
        filters=tf.reshape(vertical_kernel_2d, [3,3,1,1]),
        strides=[1,1,1,1],
        padding='SAME'
    )
    gray_hori_output = tf.nn.conv2d(
        input=tf.reshape(gray_input_tensor, [1,gray_input_shape[0],gray_input_shape[1],1]),
        filters=tf.reshape(horizontal_kernel_2d, [3,3,1,1]),
        strides=[1,1,1,1],
        padding='SAME'
    )
    
    # Converting the outputs into numpy arrays for plotting
    clr_v_out_img = tf.squeeze(clr_vert_output).numpy()
    clr_h_out_img = tf.squeeze(clr_hori_output).numpy()
    gray_v_out_img = tf.squeeze(gray_vert_output).numpy()
    gray_h_out_img = tf.squeeze(gray_hori_output).numpy()
    
    # Re-adjusting the axes
    clr_v_out_img = np.moveaxis(clr_v_out_img, 0, -1)
    clr_h_out_img = np.moveaxis(clr_h_out_img, 0, -1)
    
    outputs = [clr_vert_output, clr_hori_output, gray_vert_output, gray_hori_output]
    output_titles = ['Color vertical edges','Color horizontal edges','Gray vertical edges','Gray horizontal edges']
    
    # Formatting the images for plotting
    all_imgs = [input_arr, clr_v_out_img, clr_h_out_img, gray_input_arr, gray_v_out_img, gray_h_out_img]
    all_imgs_titles = ['Color input'] + output_titles[0:2] + ['Gray input'] + output_titles[2:]
    
    mul_plots(all_imgs, all_imgs_titles)
```

```python
plot_boundary('cube')
```
